## 3. ìŠ¤íƒ êµ¬í˜„

ì´ì œ ì†ìœ¼ë¡œ í•´ë³¸ ê²ƒì„ í”„ë¡œê·¸ë˜ë°ìœ¼ë¡œ êµ¬í˜„í•´ë³´ì. ë¨¼ì € ìŠ¤íƒ ìë£Œêµ¬ì¡°ë¥¼ ë§Œë“¤ì–´ì•¼ í•œë‹¤.

### ìŠ¤íƒ(Stack)ì´ë€?

- **LIFO(Last In, First Out)** êµ¬ì¡°
- ë§ˆì§€ë§‰ì— ë“¤ì–´ê°„ ê²ƒì´ ë¨¼ì € ë‚˜ì˜¨ë‹¤
- ì ‘ì‹œë¥¼ ìŒ“ëŠ” ê²ƒê³¼ ê°™ì€ êµ¬ì¡°
- ì£¼ìš” ì—°ì‚°:
  - `push`: ìŠ¤íƒì— ìš”ì†Œ ì¶”ê°€
  - `pop`: ìŠ¤íƒì—ì„œ ìš”ì†Œ ì œê±°í•˜ê³  ë°˜í™˜
  - `peek` (ë˜ëŠ” `top`): ìŠ¤íƒì˜ ë§¨ ìœ„ ìš”ì†Œ í™•ì¸ (ì œê±°í•˜ì§€ ì•ŠìŒ)
  - `is_empty`: ìŠ¤íƒì´ ë¹„ì–´ìˆëŠ”ì§€ í™•ì¸
  - `size`: ìŠ¤íƒì˜ í¬ê¸° ë°˜í™˜

```python
class Stack:
    def __init__(self):
        """ìŠ¤íƒ ì´ˆê¸°í™”"""
        self.items = []
    
    def push(self, item):
        """ìŠ¤íƒì— ìš”ì†Œ ì¶”ê°€"""
        self.items.append(item)
        print(f"Push {item}: {self.items}")
    
    def pop(self):
        """ìŠ¤íƒì—ì„œ ìš”ì†Œ ì œê±°í•˜ê³  ë°˜í™˜"""
        if self.is_empty():
            raise IndexError("Stack is empty")
        item = self.items.pop()
        print(f"Pop {item}: {self.items}")
        return item

    def peek(self):
        """ìŠ¤íƒì˜ ë§¨ ìœ„ ìš”ì†Œ í™•ì¸ (ì œê±°í•˜ì§€ ì•ŠìŒ)"""
        if self.is_empty():
            raise IndexError("Stack is empty")
        return self.items[-1]

    def is_empty(self) -> bool:
        """ìŠ¤íƒì´ ë¹„ì–´ìˆëŠ”ì§€ í™•ì¸"""
        return len(self.items) == 0
    
    def size(self) -> int:
        """ìŠ¤íƒì˜ í¬ê¸° ë°˜í™˜"""
        return len(self.items)
    
    def __str__(self) -> str:
        """ìŠ¤íƒì„ ë¬¸ìì—´ë¡œ í‘œí˜„"""
        return f"Stack: {self.items}"
```

```python
# ìŠ¤íƒ í…ŒìŠ¤íŠ¸
stack = Stack()
print(f"ë¹ˆ ìŠ¤íƒì¸ê°€? {stack.is_empty()}")
print(f"ìŠ¤íƒ í¬ê¸°: {stack.size()}")

# ìš”ì†Œ ì¶”ê°€
stack.push(10)
stack.push(20)
stack.push(30)

print(f"ìŠ¤íƒ í¬ê¸°: {stack.size()}")
print(f"ë§¨ ìœ„ ìš”ì†Œ: {stack.peek()}")

# ìš”ì†Œ ì œê±°
stack.pop()
stack.pop()

print(f"ë§¨ ìœ„ ìš”ì†Œ: {stack.peek()}")
print(f"ë¹ˆ ìŠ¤íƒì¸ê°€? {stack.is_empty()}")
```

## 4. ì¤‘ìœ„â†’í›„ìœ„ ë³€í™˜ê¸° êµ¬í˜„

ì´ì œ ì†ìœ¼ë¡œ í–ˆë˜ ë³€í™˜ ê³¼ì •ì„ í”„ë¡œê·¸ë˜ë°ìœ¼ë¡œ êµ¬í˜„í•´ë³´ì. **Shunting Yard Algorithm**ì„ ì‚¬ìš©í•œë‹¤.

```python
def infix_to_postfix(infix: str) -> str:
    """
    ì¤‘ìœ„ í‘œê¸°ì‹ì„ í›„ìœ„ í‘œê¸°ì‹ìœ¼ë¡œ ë³€í™˜
    Shunting Yard Algorithm ì‚¬ìš©
    """
    # ì—°ì‚°ì ìš°ì„ ìˆœìœ„ ì •ì˜
    precedence = {
        '+': 1,
        '-': 1,
        '*': 2,
        '/': 2,
    }
    
    # ìš°ê²°í•©ì„± ì—°ì‚°ì (ê±°ë“­ì œê³±)
    right_associative = {'^', '**'}
    
    stack = Stack()  # ì—°ì‚°ì ìŠ¤íƒ
    output = []      # ê²°ê³¼ ë¦¬ìŠ¤íŠ¸
    
    # ì…ë ¥ì„ í† í°ìœ¼ë¡œ ë¶„ë¦¬ (ê³µë°± ê¸°ì¤€)
    tokens = infix.split()
    
    for token in tokens:
        print(f"\nì²˜ë¦¬ ì¤‘ì¸ í† í°: {token}")
        
        # ìˆ«ìë‚˜ ë³€ìˆ˜ì¸ ê²½ìš°
        if token.isalnum():  # ìˆ«ìë‚˜ ë¬¸ì
            output.append(token)
            print(f"í”¼ì—°ì‚°ì ì¶”ê°€: {output}")
        
        # ì—¬ëŠ” ê´„í˜¸
        elif token == '(':
            stack.push(token)
        
        # ë‹«ëŠ” ê´„í˜¸
        elif token == ')':
            # ì—¬ëŠ” ê´„í˜¸ê¹Œì§€ ëª¨ë“  ì—°ì‚°ìë¥¼ pop
            while not stack.is_empty() and stack.peek() != '(':
                output.append(stack.pop())
            
            if not stack.is_empty():
                stack.pop()  # ì—¬ëŠ” ê´„í˜¸ ì œê±°
            print(f"ê´„í˜¸ ì²˜ë¦¬ í›„ ì¶œë ¥: {output}")
        
        # ì—°ì‚°ì
        elif token in precedence:
            # í˜„ì¬ ì—°ì‚°ìë³´ë‹¤ ìš°ì„ ìˆœìœ„ê°€ ë†’ê±°ë‚˜ ê°™ì€ ì—°ì‚°ìë“¤ì„ pop
            while (not stack.is_empty() and 
                   stack.peek() != '(' and
                   stack.peek() in precedence and
                   (precedence[stack.peek()] > precedence[token] or
                    (precedence[stack.peek()] == precedence[token] and token not in right_associative))):
                output.append(stack.pop())
            
            stack.push(token)
            print(f"ì—°ì‚°ì ì²˜ë¦¬ í›„ ì¶œë ¥: {output}")
    
    # ìŠ¤íƒì— ë‚¨ì€ ëª¨ë“  ì—°ì‚°ìë¥¼ pop
    while not stack.is_empty():
        output.append(stack.pop())
    
    result = ' '.join(output)
    print(f"\nìµœì¢… ê²°ê³¼: {result}")
    return result
```

```python
# ë³€í™˜ê¸° í…ŒìŠ¤íŠ¸
test_cases = [
    "2 + 3 * 4",
    "( 2 + 3 ) * 4",
    "A + B - C",
    "A * B + C / D",
    "( A + B ) * ( C - D )"
]

for infix_expr in test_cases:
    print(f"\n{'='*50}")
    print(f"ì¤‘ìœ„ì‹: {infix_expr}")
    postfix = infix_to_postfix(infix_expr)
    print(f"í›„ìœ„ì‹: {postfix}")
    print(f"{'='*50}")
```

## 5. í›„ìœ„ì‹ ê³„ì‚°ê¸° êµ¬í˜„

ì´ì œ í›„ìœ„ì‹ì„ ì‹¤ì œë¡œ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“¤ì–´ë³´ì.

```python
def evaluate_postfix(postfix: str) -> float:
    """
    í›„ìœ„ í‘œê¸°ì‹ì„ ê³„ì‚°í•˜ì—¬ ê²°ê³¼ ë°˜í™˜
    """
    stack = Stack()
    tokens = postfix.split()
    
    for token in tokens:
        print(f"\nì²˜ë¦¬ ì¤‘ì¸ í† í°: {token}")
        
        # ìˆ«ìì¸ ê²½ìš°
        if token.replace('.', '').replace('-', '').isdigit():
            stack.push(float(token))
        
        # ì—°ì‚°ìì¸ ê²½ìš°
        elif token in ['+', '-', '*', '/', '^', '**']:
            if stack.size() < 2:
                raise ValueError(f"ì—°ì‚°ì {token}ì— ëŒ€í•œ í”¼ì—°ì‚°ìê°€ ë¶€ì¡±í•©ë‹ˆë‹¤")
            
            # ì£¼ì˜: pop ìˆœì„œê°€ ì¤‘ìš”! (ì˜¤ë¥¸ìª½ í”¼ì—°ì‚°ìê°€ ë¨¼ì € popë¨)
            right_operand = stack.pop()
            left_operand = stack.pop()
            
            if token == '+':
                result = left_operand + right_operand
            elif token == '-':
                result = left_operand - right_operand
            elif token == '*':
                result = left_operand * right_operand
            elif token == '/':
                if right_operand == 0:
                    raise ValueError("0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
                result = left_operand / right_operand
            elif token in ['^', '**']:
                result = left_operand ** right_operand
            
            print(f"ê³„ì‚°: {left_operand} {token} {right_operand} = {result}")
            stack.push(result)
        
        else:
            raise ValueError(f"ì•Œ ìˆ˜ ì—†ëŠ” í† í°: {token}")
    
    if stack.size() != 1:
        raise ValueError("ì˜ëª»ëœ í›„ìœ„ í‘œê¸°ì‹ì…ë‹ˆë‹¤")
    
    final_result = stack.pop()
    print(f"\nìµœì¢… ê²°ê³¼: {final_result}")
    return final_result
```

```python
# í›„ìœ„ì‹ ê³„ì‚°ê¸° í…ŒìŠ¤íŠ¸
postfix_expressions = [
    "2 3 4 * +",      # 2 + 3 * 4 = 14
    "2 3 + 4 *",      # (2 + 3) * 4 = 20
    "5 2 - 3 * 4 +",  # (5 - 2) * 3 + 4 = 13
    "15 7 1 1 + - * 3 /",  # ë³µì¡í•œ ì˜ˆì œ
]

for postfix_expr in postfix_expressions:
    print(f"\n{'='*50}")
    print(f"í›„ìœ„ì‹: {postfix_expr}")
    try:
        result = evaluate_postfix(postfix_expr)
        print(f"ê³„ì‚° ê²°ê³¼: {result}")
    except Exception as e:
        print(f"ì˜¤ë¥˜: {e}")
    print(f"{'='*50}")
```

## 6. ì¢…í•©: ê³„ì‚°ê¸° í”„ë¡œê·¸ë¨

ì´ì œ ëª¨ë“  êµ¬ì„±ìš”ì†Œë¥¼ í•©ì³ì„œ ì™„ì „í•œ ê³„ì‚°ê¸°ë¥¼ ë§Œë“¤ì–´ë³´ì.

```python
def calculator(expression: str) -> float:
    """
    ì¤‘ìœ„ í‘œê¸°ì‹ì„ ì…ë ¥ë°›ì•„ ê³„ì‚° ê²°ê³¼ë¥¼ ë°˜í™˜í•˜ëŠ” ê³„ì‚°ê¸°
    """
    try:
        print(f"ì…ë ¥ ìˆ˜ì‹: {expression}")
        
        # 1ë‹¨ê³„: ì¤‘ìœ„ì‹ì„ í›„ìœ„ì‹ìœ¼ë¡œ ë³€í™˜
        print("\n=== 1ë‹¨ê³„: ì¤‘ìœ„ì‹ â†’ í›„ìœ„ì‹ ë³€í™˜ ===")
        postfix = infix_to_postfix(expression)
        
        # 2ë‹¨ê³„: í›„ìœ„ì‹ ê³„ì‚°
        print("\n=== 2ë‹¨ê³„: í›„ìœ„ì‹ ê³„ì‚° ===")
        result = evaluate_postfix(postfix)
        
        return result
    
    except Exception as e:
        print(f"ê³„ì‚° ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return None
```

```python
# ì™„ì „í•œ ê³„ì‚°ê¸° í…ŒìŠ¤íŠ¸
test_expressions = [
    "2 + 3 * 4",
    "( 2 + 3 ) * 4",
    "10 - 2 * 3",
    "( 10 - 2 ) * 3",
    "2 + 3 * 4 - 5",
    "( 2 + 3 ) * ( 4 - 1 )",
]

for expr in test_expressions:
    print(f"\n{'='*60}")
    result = calculator(expr)
    if result is not None:
        print(f"\nğŸ¯ '{expr}' = {result}")
    print(f"{'='*60}")
```
