## 1. 추가와 삭제: 주머니 관리하기
가장 많이 사용되는 함수들입니다. 데이터를 맨 뒤에 붙이거나, 원하는 위치에 쏙 집어넣을 수 있습니다.

- **`.append(값)`**: 리스트의 맨 마지막에 새로운 데이터를 추가합니다. (가장 많이 씀!)
- **`.insert(위치, 값)`**: 원하는 번호(인덱스) 자리에 데이터를 끼워 넣습니다.
- **`.remove(값)`**: 리스트에서 특정 값을 찾아 삭제합니다.
- **`.pop()`**: 맨 마지막 데이터를 밖으로 꺼내며 리스트에서 제거합니다.

```python
#1. 리스트를 만들고 리스트의 맨 끝(오른쪽)에 데이터 삽입하기

arr = [1,2,3]
arr.append(4)   # arr라는 리스트의 맨 끝(오른쪽)에 데이터 4를 추가해라
print(arr)
```

## 2. 정렬: 깔끔하게 줄 세우기
데이터가 뒤죽박죽일 때 순서대로 정렬하는 방법입니다.

- **`.sort()`**: 작은 것부터 큰 순서대로(오름차순) 정렬합니다.
- **`.sort(reverse=True)`**: 큰 것부터 작은 순서대로(내림차순) 정렬합니다.
- **`.reverse()`**: 현재 순서를 그대로 뒤집습니다. (정렬과는 다름!)

```python
#6. sort() 함수를 활용하여 정렬시키기
arr=[4,3,2,6,8,1]


arr.sort()   ## arr 배열을 정렬시키기 ( 오름차순 )
print(arr)  1

arr.sort(reverse = True)  ## arr 배열을 내림차순으로 정렬시키기
print(arr)

arr.reverse()   ## arr 배열의 데이터 순서를 뒤집기
print(arr)

```

## 3. 정보 함수: 리스트 속 들여다보기
- **`len(리스트)`**: 리스트에 데이터가 총 몇 개 들어있는지 알려줍니다.
- **`.count(값)`**: 특정 값이 리스트에 몇 개 있는지 세어줍니다.
- **`.index(값)`**: 특정 값이 몇 번 인덱스에 있는지 위치를 찾아줍니다.

```python
#7. len() 함수를 활용 배열의 길이(데이터의 개수)를 구하기

arr = [4,3,2,6,8,1]
a = len(arr)   ## arr 배열의 길이를 구해서 변수 a에 넣기
print(a)
arr.append(8)
print(len(arr))
```

```python
#8. count() 함수를 활용 배열속 데이터의 개수 구하기

arr = [4,3,2,6,8,1,2,2,3,2]
print(arr.count(len(arr)-7))   # arr 배열 안에 데이터 2가 몇개 있는지 확인해줘

```

```python
#9. index() 함수를 활용 배열속 데이터가 몇번방에 있는지 확인하기

arr = [4,2,2,6,8,1,4,2,5 ]
arr.reverse()
n=arr.index(6)    ## 배열 arr 속에 데이터 2가 몇번방에 있는지 알려줘
                ## 같은 값이 있다면 가장 좌측( 방번호가 빠른것) 을 알려줌
```
