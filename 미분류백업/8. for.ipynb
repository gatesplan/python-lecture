## 2. range(): 숫자 생성기
바구니를 직접 만들기 귀찮을 때, 특정 횟수만큼 반복하고 싶을 때 사용합니다.

- `range(stop)` : 0부터 (stop-1)까지 생성
- `range(start, stop)` : start부터 (stop-1)까지 생성
- `range(start, stop, step)` : start부터 (stop-1)까지 step만큼 건너뛰며 생성

- **`range(10)`**: 0, 1, 2, ..., 9 (10번 반복)
- **`range(1, 11)`**: 1, 2, 3, ..., 10 (시작과 끝 지정, 끝은 미포함!)
- **`range(1, 11, 2)`**: 1, 3, 5, 7, 9 (2씩 건너뛰기)

##  for문: 데이터 바구니에서 하나씩 꺼내기
파이썬의 `for`문은 매우 직관적입니다.
## "데이터가 담긴 주머니(Iterables)에서 알맹이를 하나씩 꺼내어 변수에 담고,
## 더 이상 꺼낼 게 없을 때까지 코드를 실행"합니다.

### ★ 핵심 문법
```python
for 변수 in 바구니(튜플, 리스트, 딕셔너리, 문자열):
    실행할 코드



````python
# [예제 1] 우주선 탑승 명단 출력 (튜플 활용)

crews = ("Captain", "Doctor", "Engineer")
for i in crews:
    print(i)
````

````python
# [예제 2] 문자열 한 글자씩 분해하기

message = "2367"
for i in message:
    print(i)

````

## 반복되는 횟수를 N으로 직접 입력받아서 코드 작성해보기

````python
## 반복횟수를 지정하지 않고 입력을 받을 수 있습니다

n=int(input("몇번 반복할까요?")) # 반복횟수 n에 입력받기  5

for i in range(1,n+1,1):         # 1부터 n까지 1씩 증가하기
  print("사랑해")
  print("나두")
  i=i+1
````

````python
## [ 실전 문제 1 ]
## n 부터 m 까지 k씩 증가하면서 사랑해를 출력하는 코드를 작성해보세요
## n 과 m 과 k를 input 함수를 사용하여 입력받습니다
m = int(input())
n = int(input())
k = int(input())


for d in range(m, n+1, k):
  print("사랑해"*d)
````

````python
# 1부터 10까지 합계를 구해보는 코드 예제
sum=0    # 합계를 구할 방을 만들고 초기화 시켜준다

for i in range (1, 11, 1):             # i의 값이 1부터 10까지 1씩 증가한다
  sum+=i                                # sum 에다가 i 값을 누적시킨다

print(sum)
````

````python
# [ 실전문제 2 ]
# 1부터 N까지의 숫자들중에 홀수의 합과 짝수의 합을 출력해주는 코드를 작성해보세요

n = int(input())

hol = zzak = 0

for i in range(1,11,1):
  if i%2 == 0:
    zzak += i
  else:
    hol += i

print(hol, zzak)
````

## 딕셔너리 반복문: Key와 Value 꺼내기
딕셔너리는 '키(Key)'와 '값(Value)'이 쌍으로 들어있어서 반복문 활용도가 매우 높습니다.
- `for k in dict:` -> 키만 꺼내기
- `for v in dict.values():` -> 값만 꺼내기
- `for k, v in dict.items():` -> 키와 값을 동시에 꺼내기 (가장 많이 사용!)

````python
# [예제] 행성별 중력 데이터 출력

planets = {"수성": 3.7, "지구": 9.8, "화성": 3.7, "목성": 24.7}

for i,j in planets.items():
    print(i)
    print(j)

````

````python
# [문제 4] 한 주간 기록된 온도가 담긴 튜플입니다.
# temps = (20, 22, 25, 21, 19, 23, 24)
# 모든 온도의 합계를 구하여 출력하세요. (반복문 사용)


temps = (20, 22, 25, 21, 19, 23, 24)
total = 0 # 합계를 저장할 변수


# 여기에 코딩하세요.
for i in temps:
  total += i


print(total)
````
