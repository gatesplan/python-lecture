# 0. 문자와 변수를 내 마음대로! f-스트링 (f-string)
우리가 지금까지 `print()`를 사용할 때, 글자와 변수를 함께 출력하려면 쉼표(`,`)나 더하기(`+`) 기호를 사용해야 해서 코드가 길어지고 헷갈리기 쉬웠습니다.

파이썬에서는 이를 아주 간단하게 해결해 주는 **f-스트링(f-string)**이라는 마법 같은 기능을 제공합니다!

### ✨ f-스트링 사용법
문자열을 만드는 따옴표 앞쪽에 알파벳 **`f`**를 붙여주기만 하면 됩니다. 그리고 문자열 안에서 변수의 값을 넣고 싶은 자리에 중괄호 `{ }`를 열고 변수 이름을 적어주면 끝입니다!

* **기존 방식:** `print("내 이름은", name, "이고, 나이는", age, "살입니다.")`
* **f-스트링:** `print(f"내 이름은 {name}이고, 나이는 {age}살입니다.")`

```python
name = "파이썬"
age = 30

# 기존 방식 (쉼표 사용)
print("안녕! 내 이름은", name, "이고, 나이는", age, "살이야.")

# f-스트링 방식 (훨씬 깔끔하죠?)
print(f"안녕! 내 이름은 {name}이고, 나이는 {age}살이야.")
```

# 1. 함수(Function)란 무엇일까?
프로그래밍을 할 때 특정 작업을 반복적으로 수행해야 할 때가 있습니다. 이때 **특정 기능을 수행하는 코드를 따로 묶어 이름을 붙여둔 것**을 '함수'라고 합니다.

* **함수의 장점**
  1. 동일한 기능이 필요한 곳에서 다시 사용할 수 있어 **코드의 중복이 줄어듭니다.**
  2. 특정 기능을 변경해야 할 때 해당 함수만 수정하면 되기 때문에 **유지보수가 편리합니다.**
  3. 코드가 의미 있는 단위로 분할되어 간결해지고, 내부 구조를 몰라도 쉽게 사용할 수 있습니다.

### ⚙️ 함수의 3가지 종류
1. **내장 함수(Built-in):** 파이썬이 기본적으로 제공하는 함수 (예: `print()`, `input()`, `len()`, `abs()`, `max()`, `divmod()`)
2. **외장 함수:** `import` 명령어로 외부 모듈이나 패키지를 불러와 사용하는 함수 (예: `random.randint()`, `matplotlib.pyplot`)
3. **사용자 정의 함수:** 프로그래머가 필요에 의해 직접 이름을 붙여 만든 함수


```python
# 내장 함수 예시
print("절댓값 구하기:", abs(-15))
print("몫과 나머지 구하기:", divmod(10, 3))
print("최댓값 구하기:", max([1,2,3,4,5]))
print("최소값 구하기:", min([1,2,3,4,5]))
print("합계구하기:", sum([1,2,3,4,5]))

# 외장 함수 예시 (random 모듈 불러오기)
import random
print("1~10 사이의 무작위 숫자 뽑기:", random.randint(1, 10))
```

### 📍 [예제] 1부터 입력받은 수까지의 합 구하기

```python
# 1. 함수 정의하기 (만들기)
def total(number):
    sum = 0
    for i in range(1, number + 1):
        sum += i
    return sum  # 계산된 sum 값을 반환


# 2. 함수 호출하기 (사용하기)
num = int(input('숫자를 입력하세요: '))
result = total(num)
print(f"1부터 {num}까지의 합은 {result}입니다.")
```

# 3. 지역 변수와 전역 변수
함수를 만들 때 주의해야 할 점이 있습니다. 바로 변수의 '활동 범위(Scope)'입니다.

* **지역 변수 (Local Variable):**
  * 함수 내부에서 선언된 변수입니다.
  * 선언된 함수 안에서만 사용할 수 있으며, 함수가 종료되면 메모리에서 사라집니다.
* **전역 변수 (Global Variable):**
  * 함수 밖에서 선언된 변수입니다.
  * 모든 영역에서 값을 불러올 수 있습니다. 단, 함수 내부에서 전역 변수의 값을 직접 변경하려면 `global` 키워드를 붙여주어야 합니다.

```python
score = 100 # 전역 변수

def change_score():
    global score # 전역 변수 score를 함수 안에서 수정하겠다고 선언!
    score -= 10
    local_bonus = 50 # 지역 변수 (이 함수 안에서만 살아있음)
    print("함수 내부 실행 완료")



change_score()
print("현재 점수:", score)
# print(local_bonus) # 주석을 해제하고 실행하면 에러가 발생합니다. (지역 변수이기 때문)
```

# 4. 종합 예제: 분식집 주문 프로그램
딕셔너리, 무한 루프, 그리고 여러 개의 함수를 조합하여 그럴싸한 분식집 메뉴 주문 프로그램을 만들 수 있습니다. 코드를 실행해 보고 어떻게 동작하는지 분석해 보세요!


```python
total = 0  # 총액을 저장할 전역 변수
menu_pan = {'떡볶이': 5000, '라면': 3000, '순대': 4000, '콜라': 1500}

def show_menu():
    print('====<메뉴판>====')
    for menu, price in menu_pan.items():
        print(f'{menu}: {price}원')
    print()

def order():
    global total  # 전역 변수 total의 값을 변경하기 위해 선언
    while True:
        menu = input('메뉴를 선택해 주세요.(주문 끝은 "n" 입력): ')
        if menu == 'n':
            break
        if menu in menu_pan:
            print(f'{menu} 주문하셔서 {menu_pan[menu]}원 추가됩니다.')
            total += menu_pan[menu]
        else:
            print('없는 메뉴입니다.')

def pay():
    print(f'\n주문하신 금액은 총 {total}원입니다.\n')

# 메인 프로그램 시작
while True:
    print('1. 메뉴판 보기\n2. 주문하기\n3. 결제하기\n4. 이용 종료\n')
    a = int(input('무엇을 하시겠습니까? '))

    if a == 1:
        show_menu()
    elif a == 2:
        order()
    elif a == 3:
        pay()
    elif a == 4:
        print('이용해 주셔서 감사합니다.')
        break
```

### 🔍 [문제 7] 약수 개수 탐지기
교재에 등장했던 코드입니다. 어떤 숫자를 넣었을 때, 그 숫자의 약수가 총 몇 개인지 계산해서 되돌려주는 함수를 완성해 보세요.


```python
#  [문제 7] 약수 개수 탐지기

# 함수를 정의하는 키워드를 적으세요.
def divisorCount(n):
    tot = 0
    for i in range(1, n + 1):
        if n % i == 0:
            tot += 1
    # 계산된 tot 값을 호출한 곳으로 되돌려주는 키워드를 적으세요.
    return tot

print("10의 약수 개수:", divisorCount(10))
```

다음 코드를 실행하기 전, 화면에 어떤 숫자가 출력될지 머릿속으로 먼저 예측해 보세요. 함수 안의 변수와 밖의 변수는 이름이 같아도 서로 다릅니다!


```python
#  [문제 10] 변수의 생존 범위 (지역 vs 전역) 분석하기

apple = 10 # 전역 변수

def eat_apple():
    global apple
    apple = 3 # 함수 내부에서 선언된 지역 변수
    print("함수 안의 사과 개수:", apple)

eat_apple()

# 함수 밖에서 apple을 출력하면 어떤 값이 나올까요?
print("함수 밖의 사과 개수:", apple)
```
