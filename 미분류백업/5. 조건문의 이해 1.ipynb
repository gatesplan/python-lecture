# [잠깐!] 비교 연산자 복습

조건식에는 주로 값을 비교하는 연산자가 사용됩니다.:



*   a > b: a가 b보다 크다
*   a < b: a가 b보다 작다
*   a == b: a와 b가 같다 (주의: =는 값을 넣는 것, ==는 같은지 물어보는 것!)
*   a != b: a와 b가 같지 않다











```python
# 실행 예제 1 - 숫자의 크기 비교하기
# 사용자로부터 숫자를 입력받아 10보다 크면 "우와! 큰 숫자네요 !" 라고 출력해봅시다
# 사용자에게 숫자를 입력받아 정수로 변환합니다.

n = int(input("숫자를 입력하세요: "))

if n > 10:
    print("우와! 큰 숫자네요!") # n이 10보다 클 때만 실행됩니다.

print("프로그램을 종료합니다.") # 조건과 상관없이 항상 실행됩니다.

```

```python
# 실행 예제 2
# 숫자를 입력받아서 10보다 크면 "우와" 라고 출력하기
# 결과에 상관없이 무조건 "사랑해" 출력하기

n=int(input("숫자를 입력하세요"))  15를 입력
if(n>10):         # 반드시 : 표시를 해주어야 합니다 !!
  print("우와")    # 참일 경우 실행할 문장은 들여쓰기를 해줘야 합니다 !!

print("사랑해")    # 조건과 상관없이 실행되기 위해서는 들여쓰기를 하지 않습니다

```

### [문제 4] 럭키 넘버 판별
사용자로부터 정수를 하나 입력받습니다.
이 숫자가 **'3의 배수이거나 5의 배수'** 이면 "럭키 넘버!"를 출력하고,

그렇지 않으면 "일반 넘버"를 출력하세요.

```python
# [문제 4] 코딩 영역
num = int(input("숫자를 입력하세요: "))

# 여기에 코드를 작성하세요.

if num%3 == 0 or num%5 == 0:
  print("럭키 넘버!")

else:
  print("일반넘버")
```

### [문제 5] 사탕 공장 QC(품질관리) 알고리즘 (
사탕의 무게(gram)를 입력받습니다. 사탕의 정상 범위는 **10g 이상 20g 이하**입니다.
- **조건:** 무게가 정상 범위를 **벗어나면** "불량품(REJECT)"을 출력하세요.
- **정상 범위라면:** "합격(PASS)"을 출력하세요.
- **주의:** 반드시 `or` 연산자를 사용하여 **범위를 벗어난 조건**을 먼저 작성해 보세요!

```python
# [문제 5] 코딩 영역
weight = float(input("사탕 무게(g): "))

# 여기에 코드를 작성하세요.

if 10 <= weight and 20 >= weight:
  print("합격(PASS)")

else:
  print("불량품(REJECT)")
```

### [문제 6] 포탈 입장 보안 시스템
4차원 세계로 가는 포탈에 입장하기 위해서는 까다로운 조건을 만족해야 합니다.
사용자로부터 **나이(age)**와 **보유 에너지(energy)**를 입력받습니다.

- **입장 조건:** (나이가 15세 이상) **이거나** (보유 에너지가 100 초과)여야 합니다.
- **단, 예외 조건:** 위 조건을 만족하더라도 **나이가 100세 이상이면** 위험하므로 절대 입장할 수 없습니다.

- **출력:** 입장이 가능하면 "포탈 가동!", 불가능하면 "입장이 거부되었습니다."를 출력하세요.
- **힌트:** `if` 안에 또 다른 `if`를 넣는 **중첩 조건문**을 활용해 보세요.

```python
# [문제6 ] 코딩 영역
age = int(input("나이: "))
energy = int(input("에너지 수치: "))

# 여기에 코드를 작성하세요.

if age >= 15 and energy > 100:
  if age <= 100:
    print("포탈 가동!")

else:
  print("입장이 거부되었습니다.")
```
