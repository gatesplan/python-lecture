### ① 슬라이싱을 이용한 일괄 변경
리스트의 특정 구간을 한 번에 다른 값들로 갈아끼울 수 있습니다.

```python
# [예제 1] 구간 수정
nums = [1, 2, 3, 4, 5]
nums[1:4] = [20, 30, 40]
print("구간 수정 결과:", nums) # [1, 20, 30, 40, 50]



# [예제 2] 리스트 복사 (주의!)
# b = a 라고 하면 주소값이 복사되어 하나를 바꾸면 둘 다 바뀝니다.
# 완전히 새로운 복사본을 만들려면 슬라이싱[:]을 쓰세요.
original = [1, 2, 3]
copy_list = original[:]
copy_list[0] = 99
print("원본:", original)    # [1, 2, 3] 유지
print("복사본:", copy_list) # [99, 2, 3]
```

```python
# [예제 2] 리스트 복사를 슬라이싱으로 안할 경우 (주의!)
original = [1, 2, 3]
copy_list = original
copy_list[0] = 99
print("원본:", original)    # [1, 2, 3] 유지
print("복사본:", copy_list)
```

## 빈 리스트 score를 만들고, 학생들의 점수를 입력받아 관리해보세요

### 1. 학생 점수를 5번 입력받아 score 리스트에 추가
### 2. 입력이 끝나면 리스트 전체를 출력

### 3. 최고점수와 최저 점수를 구해서 출력하세요

### 4. 최저 점수를 리스트에서 삭제하고 전체 합계를 구해보세요

```python
# 이곳에 코드를 작성하세요
# 힌트 : 넣는 값의 패턴을 찾아보세요
list_3 = []

for i in range(5):
  a = int(input())
  list_3.append(a)

print(list_3)

list_3.sort(reverse = True)

print(list_3[0])
print(list_3[-1])

list_3.remove(list_3[-1])

tot = 0

for i in list_3:
  tot = tot + i

print(tot)
```

## 3. 2차원 리스트 전체 훑기
표 전체의 데이터를 하나씩 꺼내려면 **for문 안에 for문**을 넣는 '중첩 반복문'이 필요합니다.

```python
# [예제 ] 모든 점수 출력하기

scores = [
    [80, 90, 85],  # 0번 학생 (행)
    [70, 65, 75],  # 1번 학생 (행)
    [100, 95, 90]  # 2번 학생 (행)
]
for row in scores:        # 각 학생(행)을 하나씩 꺼내서
    for score in row:     # 그 학생의 과목별 점수(열)를 다시 하나씩 꺼냄
        print(score, end=" ")
    print() # 한 학생이 끝나면 줄바꿈
```

```python
# [문제 2] 다음은 2명의 학생의 국어, 영어 점수입니다.
# student_scores = [[80, 70], [90, 100]]
# 반복문을 사용하여 각 학생의 '점수 합계'를 구하여 출력하세요.
# 출력 예시: 0번 학생의 합계: 150

student_scores = [[80, 70], [90, 100]]

tot1 = 0
tot2 = 0

for j in range(2):
  stu1 = student_scores[0][j]
  tot1 = tot1 + stu1

for j in range(2):
  stu2 = student_scores[1][j]
  tot2 = tot2 + stu2

print(tot1)
print(tot2)
```

```python
# [문제 3] 빈 리스트(None으로 초기화된 3x3)에
# 행 번호와 열 번호가 같은 칸( [0][0], [1][1], [2][2] )에만 1을 넣고
# 나머지는 0을 넣어 출력하세요. (반복문 활용)

matrix = [[None, None, None], [None, None, None], [None, None, None]]

for i in range(3):
  for j in range(3):
    if i == j:
      matrix[i][j] = 1
    else:
      matrix[i][j] = 0
print(matrix)
```
