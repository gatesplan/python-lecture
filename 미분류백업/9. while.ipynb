## 1. 반복문의 양대 산맥: 어떤 것을 쓸까요?

보통 **횟수**가 정해져 있을 때는 `for`문을, **특정 조건**이 유지되는 동안 계속해야 할 때는 `while`문을 사용합니다.

| 구분 | for 반복문 | while 반복문 |
| :--- | :--- | :--- |
| **핵심 개념** | **횟수** 중심 (Range, 바구니 기반) | **조건** 중심 (참/거짓 기반) |
| **주요 상황** | 10번 반복해, 리스트 요소 다 꺼내줘 | "그만"이라고 할 때까지, 에너지가 0이 될 때까지 |
| **특징** | 반복 횟수를 미리 알고 시작함 | 반복 횟수를 미리 알 수 없을 때 유리함 |

### ★ while문의 기본 구조
```python
i = 0                # ① 초기식 (시작점)
while i < 10:        # ② 조건식 (종료 조건)
    print("반복 중")  # ③ 반복 실행할 코드
    i = i + 1        # ④ 증감식 (변화가 없으면 무한 루프!)

# for문과 while문으로 작성한 코드에 대해서 비교해보세요

````python
## 사랑합니다를 세번 출력하는 for 반복문

for i in range(1,4,1):        ## (① 초기값 ② 종료조건 ③ 증감값)
  print("사랑합니다")         ##  ④ 반복실행 명령어
````

````python
## 사랑합니다를 세번 출력하는 while 반복문

i=1                           ## ① 초기값
while i<=3:                   ## ② 종료조건
  print("사랑합니다")         ## ④ 반복실행 명령어
  i=i+1                       ## ③ 증감값
````

````python
# [예제 1] 사용자에게 입력받은 숫자(n)까지 출력하기

n = int(input("어디까지 출력할까요? : "))

i = 1
while i <= n:
    print(i, end=" ")
    i += 1
````

## 2. while문의 진가: "끝을 모를 때"
프로그램을 만들다 보면 사용자가 언제 종료할지 모르는 경우가 많습니다.
## 이때 `while True`라는 **무한 루프**를 만들고,
## 특정 조건에서 `break`를 사용해 탈출하는 방식이 매우 유용합니다.

- **실제 사례:** - ATM기에서 '종료' 버튼을 누를 때까지 계속 메뉴 보여주기
    - 게임에서 캐릭터의 HP가 0이 되어 'Game Over'가 될 때까지 전투 지속하기

## 무한 루프를 활용한 while 반복문

image.png

````python
# [예제 2] 숫자를 계속 입력받다가, 총합이 100을 넘으면 종료하기
total = 0

while True:
    # num = int(input("더할 숫자를 입력하세요 (현재 합계: " + str(total) + ") : "))
    num = int(input("더할 숫자를 입력하세요  : "))
    total += num

    if total > 100:
        print("합계가 100을 넘었습니다. 프로그램을 종료합니다.")
        break  # 무한 루프를 깨고 나가는 명령어

print("최종 합계:", total)

````

````python
# [문제 9] 컴퓨터가 정한 숫자(answer = 7)를 맞출 때까지
# 계속 숫자를 입력받는 프로그램을 만드세요.
# 맞추면 "정답입니다!"를 출력하고 break로 탈출하세요.

answer = 7

while True:
  n = int(input("숫자 맞춰봐 : "))

  if n == answer:
    print("정답입니다!")
      break
````

````python
# [문제 10] 사용자로부터 먹고 싶은 재료를 계속 입력받아 튜플이나 리스트처럼 보여주는 시뮬레이션입니다.
# 사용자가 "끝"이라고 입력할 때까지 재료를 입력받으세요.
# (출력 예시: "베이컨을(를) 추가합니다!")

while True:
  n = input("먹고 싶은 재료를 입력하세요 : ")

  if n == "끝":
    print("배부르다!")
    break

  print(n, "을(를) 추가합니다!")
````

````python
# [문제 11] 합계가 30점이 넘을 때 까지 입력받은 점수를 누적하는 코드를 작성하는데,
## 점수를 총 몇번 입력했는지를 알려주는 코드를 작성해보세요.
c = 0
total = 0

while True:
  n = int(input("숫자를 입력하세요 "))
  total = n + total
  c += 1

  if total >= 30:
    print(c, total)
    break
````

##  continue: "이번 차례만 건너뛰기"
반복문 안에서 `continue`를 만나면, 그 아래에 있는 코드들은 실행하지 않고 **즉시 반복문의 처음(조건 검사 부분)으로 돌아갑니다.**

- **break**: 반복문을 아예 **종료**하고 밖으로 나감 (탈출)
- **continue**: 반복문은 계속하되, 현재 순서만 **생략**하고 다음 순서로 넘어감 (스킵)

### 💡 continue는 언제 유용할까요?
특정 조건에 해당하는 데이터를 **제외**하고 처리해야 할 때 유용합니다.
- **예시 1:** 학생들의 성적을 처리하다가 '결석'한 학생만 빼고 평균을 낼 때
- **예시 2:** 100층까지 있는 엘리베이터에서 '4층'과 '13층'만 안 서고 지나갈 때
- **예시 3:** 게임에서 '무적 상태'인 적에게는 데미지 계산을 건너뛸 때

````python
# [예제 1] 1부터 10까지 숫자 중 짝수는 건너뛰고 홀수만 출력하기
i = 0
while i < 10:
    i += 1
    if i % 2 == 0:  # 짝수라면?
        continue    # 아래 print를 실행하지 않고 다시 위(while 조건문)로 올라감
    print(i, end=" ")

````

````python
##[예제 2]  break 의 반대로 continue 명령어를 수행하여 뛰어넘기가 가능합니다

i=0
while i<3:
  print("사랑합니다")
  i=i+1

  if i%2==0:
    continue

  print("나두용")
````

````python
# [문제 12] 식당 예약 명단 튜플이 있습니다.
# 중간에 "노쇼(No-Show)"라고 적힌 예약은 건너뛰고
# 정상 예약 손님의 이름만 출력하세요.

reservations = ("김철수", "노쇼", "이영희", "박지성", "노쇼", "손흥민")
i = 0

while i <= 6:
  if reservations[i] == "노쇼":
    i += 1
    continue

  else:
    print(reservations[i])
    i += 1
````
