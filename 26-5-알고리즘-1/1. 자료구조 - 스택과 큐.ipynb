# 1. 자료구조 - 스택과 큐

## 이 노트북에서 배우는 것

- **스택(Stack)**: 나중에 넣은 것이 먼저 나온다. LIFO (Last In First Out)
- **큐(Queue)**: 먼저 넣은 것이 먼저 나온다. FIFO (First In First Out)

## 왜 한 가지를 세 가지 방식으로 만드나?

같은 스택/큐를 **세 가지 코딩 방식**으로 각각 만들어 본다.

| 방식 | 자료와 동작 | 정식 명칭 |
|------|-----------|-----------|
| A. 전역 리스트 + 함수 | 분리 (전역 공유) | 절차적 |
| B. 리스트 전달 + 함수 | 분리 (인자로 전달) | 절차적 |
| C. 클래스 | 결합 (캡슐화) | 객체지향 |

목표는 두 가지다.

1. **짜보기**: 여러 방식으로 직접 구현하며 코딩 폭을 넓힌다.
2. **읽어보기**: 남이 짠 코드를 읽고 동작을 예측한다.

시험 코드는 출제자 마음대로 어떤 방식으로든 나온다. 그래서 "내가 짤 수 있다"보다
**"남의 코드를 읽고 동작을 이해한다"** 가 더 중요하다. 각 방식마다 [읽어보기] 문제가 붙는 이유다.

## 셀 구성 방법

각 구현은 두 칸으로 나뉜다.

- **[시연]** 칸: 주석으로 어떤 함수/메서드를 만들지 시그니처만 적어 두었다. **수업 중 여기에 함께 코드를 작성**한다.
- **[정답]** 칸: 완성된 코드. 시연으로 작성한 뒤 비교/확인용.

# 스택 (Stack) - LIFO

접시를 쌓는 것과 같다. 맨 위에만 올리거나(push) 내릴 수(pop) 있다.
가장 나중에 올린 접시가 가장 먼저 내려온다.

```
push(10) -> [10]
push(20) -> [10, 20]
push(30) -> [10, 20, 30]
pop()    -> 30 반환,  [10, 20]
pop()    -> 20 반환,  [10]
```

기본 동작 이름:
- push(x): 맨 위에 넣기
- pop(): 맨 위에서 빼서 반환
- peek(): 맨 위를 빼지 않고 확인
- is_empty(): 비었는지 확인
- size(): 개수


## 방식 A: 절차적 - 전역 리스트

스택 데이터가 함수 **바깥**의 전역 변수에 있다. 함수들이 그 전역 변수를 직접 건드린다.

주의: 이 방식은 사실 **좋지 않은 방식**이다. 하지만 **교과서와 학교 시험이 이 방식을 자주 쓴다.**
그래서 채택할 방식이라기보다 **반드시 읽을 줄 알아야 하는 방식**으로 익힌다.


````python
# [시연] 전역 리스트로 스택 - 함께 작성해봅시다
# stack = []          # 전역 스택
# def push(x):        # stack 맨 뒤에 넣기
# def pop():          # stack 맨 뒤에서 빼서 반환
# def peek():         # 맨 위 확인 (빼지 않음)
# def is_empty():     # 비었는지
# def size():         # 개수
````

````python
# [정답] 방식 A: 절차적 - 전역 리스트
stack = []  # 전역 스택

def push(x):
    stack.append(x)       # 맨 뒤(위)에 넣기

def pop():
    return stack.pop()    # 맨 뒤(위)에서 빼기 - LIFO

def peek():
    return stack[-1]      # 맨 위 확인 (빼지 않음)

def is_empty():
    return len(stack) == 0

def size():
    return len(stack)
````

````python
# 사용해보기
push(10) # [10]
push(20) # [10, 20]
push(30) # [10, 20, 30]
print(stack)     # [10, 20, 30]
print(pop())     # 30  [10, 20]
print(peek())    # 20
print(size())    # 2
````

실행하기 전에 종이에 먼저 예측해 보자. 그 다음 실행해서 맞는지 확인한다.


````python
stack = []       # 스택 초기화
push(1)
push(2)
push(3)
print(pop())     # ?
push(4)
print(peek())    # ?
print(stack)     # ?
````

**정답:** `3` / `4` / `[1, 2, 4]`

- push(1,2,3) 후: [1, 2, 3]
- pop() -> 3 반환, [1, 2]
- push(4) -> [1, 2, 4]
- peek() -> 4 (빼지 않음)

### 이 방식의 문제점

전역 `stack` 하나에 묶여 있어서 **스택을 두 개 동시에 쓸 수 없다.**
그리고 코드 어디서든 `stack`을 건드릴 수 있어 실수하기 쉽다. 다음 방식이 이를 해결한다.


## 방식 B: 절차적 - 리스트를 인자로 전달

스택 데이터를 **함수에 넘겨서** 다룬다. 전역 변수에 의존하지 않는다.
자료(리스트)와 동작(함수)은 여전히 **분리**되어 있다.


````python
# [시연] 리스트 전달로 스택 - 함께 작성해봅시다
# 데이터 - 행동
# def push(stack, x):     # stack 맨 뒤에 넣기
# def pop(stack: list):
# def peek(stack: list):
# def is_empty(stack: list):
# def size(stack: list):
````

````python
# [정답] 방식 B: 절차적 - 리스트 전달
def push(stack, x):
    stack.append(x)

def pop(stack):
    return stack.pop()

def peek(stack):
    return stack[-1]

def is_empty(stack):
    return len(stack) == 0

def size(stack):
    return len(stack)
````

````python
# 사용해보기 - 스택을 두 개 만들 수 있다!
a = []
b = []
push(a, 1)
push(a, 2)
push(b, 99)
print(a)   # [1, 2]
print(b)   # [99]
print(pop(a))   # 2
print(a)        # [1]
````

참고: 리스트를 함수에 넘기면 **같은 리스트를 가리킨다**(참조). 그래서 `push` 안에서 `append` 한
결과가 바깥에도 반영된다. 그래서 push는 값을 돌려줄 필요가 없고, pop만 뺀 값을 반환한다.

### [읽어보기] 무엇이 출력될까?


````python
s1 = []
s2 = []
push(s1, "가")
push(s1, "나")
push(s2, "A")
print(pop(s1))   # ?
print(s1)        # ['가']
print(s2)        # ['A']
````

**정답:** `나` / `['가']` / `['A']`

s1과 s2가 서로 다른 리스트라서 독립적으로 동작한다. 방식 A(전역)로는 불가능했던 일이다.


## 방식 C: 객체지향 - 클래스로 묶기(캡슐화)

자료(`self.items`)와 동작(메서드)을 **하나로 묶는다.** 이것을 **캡슐화**라고 한다.
각 스택 객체는 자기만의 데이터를 가진다.

- `self`: 이 객체 자신을 가리키는 이름표
- `__init__`: 객체가 만들어질 때 처음 실행되는 부분 (데이터 준비)


````python
# [시연] 클래스로 스택 - 함께 작성해봅시다
# class Stack:
#     def __init__(self):    # 빈 리스트 self.items 준비
#     def push(self, x):     # self.items 맨 뒤에 추가
#     def pop(self):         # self.items 맨 뒤에서 빼서 반환
#     def peek(self):        # 맨 위 확인
#     def is_empty(self):    # 비었는지
#     def size(self):        # 개수

````

````python
# [정답] 방식 C: 객체지향 - 클래스
class Stack:
    def __init__(self):
        self.items = []  # 이 스택만의 데이터

    def push(self, x):
        self.items.append(x)

    def pop(self):
        return self.items.pop()

    def peek(self):
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
````

````python
# 사용해보기
s = Stack()
t = Stack()        # 또 다른 스택 (독립적)
s.push(10)
s.push(20)
s.push(30)
print(s.pop())     # 30 [10, 20]
print(s.size())    # 2

t.push("가")
print(t.items)     # ['가']
print(s.items)     # [10, 20]
````

### [읽어보기] 무엇이 출력될까?


````python
s = Stack()
for x in [5, 10, 15]:
    s.push(x)
print(s.pop() + s.pop())   # 25
print(s.size())            # 1
````

**정답:** `25` / `1`

- push 후: [5, 10, 15]
- pop() -> 15, 다시 pop() -> 10, 합은 25
- 남은 것은 [5], size는 1

### 세 방식을 지나오며

- A(전역): 제일 단순하지만 스택 하나에 묶이고 실수하기 쉽다. **시험에 나오니 읽을 줄 알아야 한다.**
- B(전달): 스택을 여러 개 쓸 수 있다. 대신 매번 리스트를 넘겨야 한다.
- C(클래스): 자료와 동작이 묶여 깔끔하다. **클래스를 왜 쓰는지**가 여기서 드러난다.


## 스택 응용

스택으로 실제 문제를 풀어 본다. (방식 C 클래스를 사용)


````python
# [시연] 스택으로 문자열 뒤집기 - 함께 작성해봅시다
def reverse_string(text):
    s = Stack()

    for c in text:
        s.push(c)

    result = ''
    while not s.is_empty():
        result += s.pop()

    return result

reverse_string('abcdef')
````

````python
# [정답] 응용 1: 문자열 뒤집기
def reverse_string(text):
    s = Stack()
    for ch in text:
        s.push(ch)
    result = ""
    while not s.is_empty():
        result += s.pop()
    return result

print(reverse_string("hello"))   # olleh
print(reverse_string("스택연습")) # 습연택스
````

````python
# [시연] 괄호 짝 검사 - 함께 작성해봅시다

def is_balanced(text):
    s = Stack()

    for c in text:
        if c=='(':
            s.push(c)
        elif c==')':
            if s.is_empty():
                return False
            s.pop()

    return s.is_empty()

is_balanced('(((((((((())))))))((()))((()))(((())')
````

````python
# [정답] 응용 2: 괄호 짝 맞는지 검사
def is_balanced(text):
    s = Stack()
    for ch in text:
        if ch == "(":
            s.push(ch)
        elif ch == ")":
            if s.is_empty():
                return False      # 닫는 괄호가 먼저 나옴
            s.pop()
    return s.is_empty()           # 안 닫힌 괄호가 없어야 True

print(is_balanced("(())"))    # True
print(is_balanced("(()"))     # False
print(is_balanced(")("))      # False
````

# 큐 (Queue) - FIFO

줄 서기와 같다. 먼저 온 사람이 먼저 나간다.

```
enqueue(10) -> [10]
enqueue(20) -> [10, 20]
enqueue(30) -> [10, 20, 30]
dequeue()   -> 10 반환,  [20, 30]
dequeue()   -> 20 반환,  [30]
```

기본 동작 이름:
- enqueue(x): 뒤에 넣기
- dequeue(): 앞에서 빼서 반환
- front(): 맨 앞을 빼지 않고 확인
- is_empty(), size()

스택과 다른 점은 **빼는 위치가 앞(index 0)** 이라는 것뿐이다.


## 방식 A: 절차적 - 전역 리스트


````python
# [시연] 전역 리스트로 큐 - 함께 작성해봅시다
# queue = []          # 전역 큐
# def enqueue(x):     # queue 뒤에 넣기
# def dequeue():      # queue 앞에서 빼서 반환
# def front():        # 맨 앞 확인
# def is_empty():     # 비었는지
queue = []
def enqueue(x):
    queue.append(x)
def dequeue():
    if len(queue) != 0:
        return queue.pop(0)
    return None
def front():
    return queue[0]
def is_empty():
    return len(queue) == 0
````

````python
# [정답] 방식 A: 전역 리스트
queue = []

def enqueue(x):
    queue.append(x)        # 뒤에 넣기

def dequeue():
    return queue.pop(0)    # 앞에서 빼기 - FIFO

def front():
    return queue[0]

def is_empty():
    return len(queue) == 0
````

````python
queue = []
enqueue("가")
enqueue("나")
enqueue("다")
print(dequeue())   # 가
print(front())     # 나
print(queue)       # ['나', '다']
````

## 방식 B: 절차적 - 리스트 전달


````python
# [시연] 리스트 전달로 큐 - 함께 작성해봅시다
# def enqueue(queue, x):  # queue 뒤에 넣기
# def dequeue(queue):     # queue 앞에서 빼서 반환
# def front(queue):       # 맨 앞 확인
# def is_empty(queue):    # 비었는지
````

````python
# [정답] 방식 B: 리스트 전달
def enqueue(queue, x):
    queue.append(x)

def dequeue(queue):
    return queue.pop(0)

def front(queue):
    return queue[0]

def is_empty(queue):
    return len(queue) == 0
````

````python
line1 = []
line2 = []
enqueue(line1, 1)
enqueue(line1, 2)
enqueue(line2, 100)
print(dequeue(line1))   # 1
print(line1)            # [2]
print(line2)            # [100]
````

### [읽어보기] 무엇이 출력될까?


````python
q = []
enqueue(q, "A")
enqueue(q, "B")
print(dequeue(q))   # ?
enqueue(q, "C")
print(dequeue(q))   # ?
print(q)            # ?
````

**정답:** `A` / `B` / `['C']`

먼저 넣은 것이 먼저 나온다(FIFO). A를 빼고, C를 넣고, 그 다음 앞에 있던 B를 뺀다. 남는 것은 C.


## 방식 C: 객체지향 - 클래스


````python
# [시연] 클래스로 큐 - 함께 작성해봅시다
# class Queue:
#     def __init__(self):     # 빈 리스트 self.items 준비
#     def enqueue(self, x):   # self.items 뒤에 넣기
#     def dequeue(self):      # self.items 앞에서 빼서 반환
#     def front(self):        # 맨 앞 확인
#     def is_empty(self):     # 비었는지
#     def size(self):         # 개수
````

````python
# [정답] 방식 C: 클래스
class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, x):
        self.items.append(x)

    def dequeue(self):
        return self.items.pop(0)

    def front(self):
        return self.items[0]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)
````

````python
# 응용: 대기열 시뮬레이션
q = Queue()
for name in ["가", "나", "다"]:
    q.enqueue(name)

while not q.is_empty():
    print(q.dequeue(), "님 입장")
````

## 심화 훅: 큐를 리스트로 만들면 느리다

`dequeue`에서 `queue.pop(0)` 을 쓰면, 맨 앞을 빼고 **나머지 전부를 한 칸씩 앞으로 당긴다.**
데이터가 n개면 매번 n번 옮기는 셈이라 느리다. (스택의 `pop()` 은 맨 뒤라서 당길 필요가 없어 빠르다.)

해결 아이디어: 실제로 지우지 말고, **머리 위치만 표시**로 옮긴다.


````python
# [시연] 머리 인덱스로 개선한 큐 - 함께 작성해봅시다
# class Queue2:
#     def __init__(self):     # self.items = [], self.head = 0
#     def enqueue(self, x):   # 뒤에 넣기
#     def dequeue(self):      # head 위치 값 반환하고 head 한 칸 옮기기
#     def is_empty(self):     # head가 끝에 도달했는지
````

````python
# [정답] 머리 인덱스로 개선한 큐
class Queue2:
    def __init__(self):
        self.items = []
        self.head = 0             # 맨 앞이 어디인지 가리키는 표시

    def enqueue(self, x):
        self.items.append(x)

    def dequeue(self):
        x = self.items[self.head]
        self.head += 1            # 지우지 않고 표시만 한 칸 옮김
        return x

    def is_empty(self):
        return self.head >= len(self.items)
````

````python
q = Queue2()
for x in [10, 20, 30]:
    q.enqueue(x)
print(q.dequeue())   # 10
print(q.dequeue())   # 20
print(q.head)        # 2  (실제 데이터는 아직 [10,20,30])
````

이 "빠르다 / 느리다" 이야기는 뒤에 나올 **시간 복잡도** 단원에서 자세히 다룬다.
지금은 "리스트로 큐를 만들면 앞에서 빼는 게 비효율적일 수 있다" 정도만 기억하면 된다.

# 정리

## 스택 vs 큐

| | 스택(Stack) | 큐(Queue) |
|---|---|---|
| 규칙 | LIFO (나중 것이 먼저) | FIFO (먼저 것이 먼저) |
| 넣기 | push (맨 뒤) | enqueue (맨 뒤) |
| 빼기 | pop (맨 뒤) | dequeue (맨 앞) |
| 비유 | 접시 더미 | 줄 서기 |

## 세 가지 구현 방식

| 방식 | 자료와 동작 | 명칭 | 특징 |
|------|-----------|------|------|
| A. 전역 리스트 | 분리 (전역 공유) | 절차적 | 단순하나 하나에 묶임. **시험에 나오니 읽을 줄 알기** |
| B. 리스트 전달 | 분리 (전달) | 절차적 | 여러 개 사용 가능. 매번 전달 |
| C. 클래스 | 결합 (캡슐화) | 객체지향 | 자료+동작을 묶어 깔끔 |

핵심: 시험 코드는 어떤 방식으로든 나올 수 있다. 세 방식을 모두 **읽고 동작을 예측**할 수 있어야 한다.
