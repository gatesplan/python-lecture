## 먼저 실행할 것
아래 **준비 셀**을 반드시 먼저 실행해야 스택/큐를 쓸 수 있다.


```python
# 준비 셀 - 스택과 큐 (1단원에서 만든 것)
class Stack:
    def __init__(self):
        self.items = []
    def push(self, x):
        self.items.append(x)
    def pop(self):
        return self.items.pop()
    def peek(self):
        return self.items[-1]
    def is_empty(self):
        return len(self.items) == 0
    def size(self):
        return len(self.items)

class Queue:
    def __init__(self):
        self.items = []
    def enqueue(self, x):
        self.items.append(x)
    def dequeue(self):
        return self.items.pop(0)
    def front(self):
        return self.items[0]
    def is_empty(self):
        return len(self.items) == 0
    def size(self):
        return len(self.items)

print("준비 완료")
```

## 문제 2. 진법 변환

10진수 `n`을 `base` 진법 문자열로 바꾸는 함수 `to_base(n, base)` 를 만들어라.
- `base`로 나눈 **나머지를 스택에 쌓았다가** 꺼내면 자리 순서가 맞는다.
- 10 이상 자리는 문자로: 10->A, 11->B, ... 15->F
- 예: `to_base(13, 2)` -> `"1101"`, `to_base(255, 16)` -> `"FF"`


```python
# [직접 풀기] n진법 변환
# def to_base(n, base):
#     digits = "0123456789ABCDEF"   # 나머지를 문자로 바꿀 때 사용
#     스택 만들기
#     n == 0 이면 "0" 반환 (예외)
#     n > 0 동안: (n % base)에 해당하는 문자를 push, n = n // base
#     스택을 pop 하며 문자열로 이어붙이기
#     결과 반환
```

```python
# [정답]
def to_base(n, base):
    digits = "0123456789ABCDEF"
    s = Stack()
    if n == 0:
        return "0"
    while n > 0:
        s.push(digits[n % base])   # 나머지를 문자로 쌓기
        n = n // base
    result = ""
    while not s.is_empty():
        result += s.pop()          # 꺼내면 역순 = 올바른 순서
    return result

print(to_base(13, 2))    # 1101
print(to_base(255, 16))  # FF
print(to_base(100, 8))   # 144
print(to_base(0, 2))     # 0
```

## 문제 3. 괄호 검사 (여러 종류)

`()`, `[]`, `{}` 세 종류 괄호가 **올바르게 짝지어졌는지** 판별하는 `is_balanced(text)` 를 만들어라.
- 여는 괄호는 스택에 쌓는다.
- 닫는 괄호를 만나면 스택 맨 위와 **짝이 맞아야** 한다. 비었거나 짝이 다르면 실패.
- 예: `"([{}])"` -> True, `"([)]"` -> False


```python
# [직접 풀기] 괄호 검사
# def is_balanced(text):
#     pairs = {')':'(', ']':'[', '}':'{'}   # 닫힘 -> 열림 짝
#     스택 만들기
#     각 글자 ch에 대해:
#         여는 괄호이면 push
#         닫는 괄호이면: 스택이 비었거나 pop한 것이 짝과 다르면 False
#     끝나고 스택이 비어 있으면 True (안 닫힌 게 없어야 함)
```

```python
# [정답]
def is_balanced(text):
    pairs = {')': '(', ']': '[', '}': '{'}
    opens = "([{"
    s = Stack()
    for ch in text:
        if ch in opens:
            s.push(ch)
        elif ch in pairs:
            if s.is_empty() or s.pop() != pairs[ch]:
                return False
    return s.is_empty()

print(is_balanced("(){}[]"))   # True
print(is_balanced("([{}])"))   # True
print(is_balanced("([)]"))     # False
print(is_balanced("(()"))      # False
```

## 문제 4. 카드 버리기 (큐)

1부터 N까지 번호 카드가 순서대로 놓여 있다(1이 맨 위).
다음을 카드가 한 장 남을 때까지 반복한다.
1. 맨 위 카드를 **버린다.**
2. 그다음 맨 위 카드를 **맨 아래로 옮긴다.**

마지막에 남는 카드 번호를 반환하는 `last_card(n)` 을 만들어라.
- 예: `last_card(6)` -> `4`


```python
# [직접 풀기] 카드 버리기
# def last_card(n):
#     큐에 1, 2, ..., n 을 순서대로 enqueue
#     카드가 2장 이상인 동안:
#         dequeue()            # 맨 위 버리기
#         옮길카드 = dequeue()  # 그다음 카드를
#         enqueue(옮길카드)     # 맨 아래로
#     남은 한 장(front) 반환
```

```python
# [정답]
def last_card(n):
    q = Queue()
    for i in range(1, n + 1):
        q.enqueue(i)
    while q.size() > 1:
        q.dequeue()               # 맨 위 버리기
        q.enqueue(q.dequeue())    # 다음 카드를 맨 아래로
    return q.front()

print(last_card(1))   # 1
print(last_card(2))   # 2
print(last_card(4))   # 4
print(last_card(6))   # 4
```

## 문제 5. (도전) 중위 -> 후위 변환

우리가 쓰는 수식(중위표기) `3 + 4 * 2` 를 후위표기 `3 4 2 * +` 로 바꿔라.
이것이 계산기가 **연산자 우선순위**를 처리하는 방법이다(Shunting-yard 알고리즘, **연산자 스택** 사용).

- 연산자는 `+ - * /`, 괄호 `( )` 포함.
- 입력은 **공백으로 구분된 토큰**이라고 가정: `"( 3 + 4 ) * 2"`
- 규칙:
  - 숫자 -> 바로 출력
  - 연산자 -> 스택 위 연산자의 우선순위가 **크거나 같으면** 꺼내 출력하고, 자신을 push
  - `(` -> push
  - `)` -> `(` 가 나올 때까지 꺼내 출력하고 `(` 는 버림
  - 끝나면 스택에 남은 연산자를 모두 꺼내 출력

### 변환 과정 예시 (왼쪽부터 한 토큰씩 처리)

**예시 1) `3 + 2`  (우선순위 다툼 없음)**

| 토큰 | 동작 | 스택 | 출력 |
|------|------|------|------|
| 3 | 숫자 -> 출력 |  | 3 |
| + | push | + | 3 |
| 2 | 숫자 -> 출력 | + | 3 2 |
| (끝) | 스택 비우기 |  | 3 2 + |

결과: `3 2 +`

**예시 2) `3 + 2 * 3`  (`*` 가 `+` 보다 우선순위 높음)**

| 토큰 | 동작 | 스택 | 출력 |
|------|------|------|------|
| 3 | 숫자 -> 출력 |  | 3 |
| + | push | + | 3 |
| 2 | 숫자 -> 출력 | + | 3 2 |
| * | +보다 우선순위 높음 -> push | + * | 3 2 |
| 3 | 숫자 -> 출력 | + * | 3 2 3 |
| (끝) | 스택 비우기(* 먼저, + 나중) |  | 3 2 3 * + |

결과: `3 2 3 * +`

**예시 3) `( 3 + 2 ) * 3`  (괄호 처리)**

| 토큰 | 동작 | 스택 | 출력 |
|------|------|------|------|
| ( | push | ( |  |
| 3 | 숫자 -> 출력 | ( | 3 |
| + | top이 '(' 라 안 꺼내고 push | ( + | 3 |
| 2 | 숫자 -> 출력 | ( + | 3 2 |
| ) | '(' 나올 때까지 꺼냄(+), '(' 버림 |  | 3 2 + |
| * | push | * | 3 2 + |
| 3 | 숫자 -> 출력 | * | 3 2 + 3 |
| (끝) | 스택 비우기 |  | 3 2 + 3 * |

결과: `3 2 + 3 *`


```python
# [직접 풀기] 중위 -> 후위
# def to_postfix(expr):
#     prec = {'+':1, '-':1, '*':2, '/':2}   # 우선순위
#     연산자 스택, 출력 리스트 준비
#     for token in expr.split():
#         숫자이면        -> 출력 리스트에 추가
#         연산자이면      -> 스택이 안 비었고 top이 '(' 아니고 prec[top] >= prec[token] 인 동안 꺼내 출력, 그 뒤 push
#         '(' 이면        -> push
#         ')' 이면        -> top이 '(' 아닐 동안 꺼내 출력, 그다음 '(' 를 pop(버림)
#     스택에 남은 것 모두 꺼내 출력
#     return " ".join(출력 리스트)
```

```python
# [정답]
def to_postfix(expr):
    prec = {'+': 1, '-': 1, '*': 2, '/': 2}
    s = Stack()
    output = []
    for token in expr.split():
        if token in prec:                       # 연산자
            while (not s.is_empty() and s.peek() != '('
                   and prec[s.peek()] >= prec[token]):
                output.append(s.pop())
            s.push(token)
        elif token == '(':
            s.push(token)
        elif token == ')':
            while s.peek() != '(':
                output.append(s.pop())
            s.pop()                             # '(' 버리기
        else:                                   # 숫자
            output.append(token)
    while not s.is_empty():
        output.append(s.pop())
    return " ".join(output)

print(to_postfix("3 + 4 * 2"))        # 3 4 2 * +
print(to_postfix("( 3 + 4 ) * 2"))    # 3 4 + 2 *
```

## 문제 6. 후위식 계산

후위표기 수식 `"3 4 2 * +"` 를 계산해 값을 구하는 `eval_postfix(expr)` 를 만들어라(**피연산자 스택** 사용).

- 입력은 공백으로 구분된 토큰.
- 숫자 -> 스택에 push
- 연산자 -> 스택에서 **두 개 꺼내** 계산 후 결과를 push
- **순서 주의:** 먼저 꺼낸 게 오른쪽 피연산자(b), 나중 꺼낸 게 왼쪽(a). `a (연산) b`
- 예: `eval_postfix("3 4 2 * +")` -> `11`   (3 + 4*2)


```python
# [직접 풀기] 후위식 계산
# def eval_postfix(expr):
#     피연산자 스택 준비
#     for token in expr.split():
#         연산자('+','-','*','/')이면:
#             b = pop()   # 오른쪽
#             a = pop()   # 왼쪽
#             a 연산 b 결과를 push
#         숫자이면:
#             int(token) 으로 바꿔 push
#     마지막 남은 값 반환
```

```python
# [정답]
def eval_postfix(expr):
    s = Stack()
    for token in expr.split():
        if token in ("+", "-", "*", "/"):
            b = s.pop()      # 오른쪽 피연산자
            a = s.pop()      # 왼쪽 피연산자
            if token == "+":
                s.push(a + b)
            elif token == "-":
                s.push(a - b)
            elif token == "*":
                s.push(a * b)
            else:
                s.push(a / b)
        else:
            s.push(int(token))
    return s.pop()

print(eval_postfix("3 4 2 * +"))   # 11
print(eval_postfix("3 4 + 2 *"))   # 14
```

### 연결해보기: 중위식을 넣으면 답이 나온다

문제 5(변환) + 문제 6(계산)을 이으면 간단한 계산기가 된다.


```python
expr = "( 3 + 4 ) * 2"
post = to_postfix(expr)
print(post, "=", eval_postfix(post))   # 3 4 + 2 * = 14
```
